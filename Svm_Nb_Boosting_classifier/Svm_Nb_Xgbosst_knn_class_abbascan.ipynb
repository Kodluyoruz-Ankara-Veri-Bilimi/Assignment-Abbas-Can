```python
df = pd.read_csv("hmelq.csv")
df.head(5)
```

```python
df.dropna(inplace = True)
```

```python
df_dummies = pd.get_dummies(df)
```

# SVC

```python
X = df_dummies.drop("bad", axis = 1)
y = df_dummies["bad"]
X_train , X_test, y_train, y_test = train_test_split(X, y , test_size = 0.25, random_state = 42)
```

```python
svc = SVC()
```

```python
svc_model = svc.fit(X_train, y_train)
y_pred = svc_model.predict(X_test)

ac = accuracy_score(y_test, y_pred)
con_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
svc_model
```

```python
svc_params = {"kernel": ["linear", 'poly', 'rbf'],
             "C": [1,2,3,6,8]}
```

```python
svc_model_cv = GridSearchCV(svc, svc_params, cv=5, n_jobs= -1, verbose= 2)
```

```python
svc_model_cv.fit(X_train, y_train)
```

```python
svc_model_cv.best_params_
```

```python
svc_model = SVC(C = 8, kernel = "linear")
svc_model_tuned = svc_model.fit(X_train, y_train)
```

```python
y_pred_tuned = svc_model_tuned.predict(X_test)

ac_tuned = accuracy_score(y_test, y_pred_tuned)
con_matrix_tuned = confusion_matrix(y_test, y_pred_tuned) 
class_report_tuned = classification_report(y_test, y_pred_tuned)


print("Accuracy Score: {} \nConfusion Matrix: \n{}\n*********************************\nClassification Report\n{}\n\n <--------------------------------------->\n\n\nAccuracy Score Tuned: {} \nConfusion Matrix Tuned: \n{}\n*********************************\nClassification Report Tuned\n{} ".format(ac,con_matrix, class_report, ac_tuned, con_matrix_tuned, class_report_tuned))
```

# XG Boost

```python
X = df_dummies.drop("bad", axis = 1)
y = df_dummies["bad"]
X_train , X_test, y_train, y_test = train_test_split(X, y , test_size = 0.25, random_state = 42)
```

```python
xgb = XGBClassifier()
```

```python
y_pred = xgb_model.predict(X_test)

ac = accuracy_score(y_test, y_pred)
con_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
```

```python
xgb_params = {
        'min_child_weight': [0.8 ,0.9, 1],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5]
        }
```

```python
df = pd.read_excel('HW_Data_Set.xlsx', index_col=None)
```

```python
target_obj_20=[]

for i in df['20_target']: 
    
    if i == 0 :
    
        target_obj_20.append(1)
    
    elif i < 0: 
    
        target_obj_20.append(0)
        
    else :
        
        target_obj_20.append(2)

df['target_obj_20'] = target_obj_20          

target_obj_50=[]

for i in df['50_target']: 
    
    if i == 0 :
    
        target_obj_50.append(1)
    
    elif i < 0: 
    
        target_obj_50.append(0)
        
    else :
        
        target_obj_50.append(2)
    
df['target_obj_50'] = target_obj_50  


target_obj_90=[]

for i in df['90_target']: 
    
    if i == 0 :
    
        target_obj_90.append(1)
    
    elif i < 0: 
    
        target_obj_90.append(0)
        
    else :
        
        target_obj_90.append(2)
    
df['target_obj_90'] = target_obj_90
```
