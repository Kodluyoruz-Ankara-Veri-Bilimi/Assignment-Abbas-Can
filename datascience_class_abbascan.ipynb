```python
class bilgi():
    
    def __init__(self):
        ...
        
    def tanimlayici_ist(self):
        # Tanımlayıcı istatistik bilgilerini bize gösterir.
        a = self.veri.describe()
        print(a)
    def on_bilgi(self):
        # Veri seti hakkında bize genel bir bilgi verir.
        a = self.head()
        b = self.dtypes
        print(a,b)
        
class duzenleme():  
    
    def __init__(self, veri):
        ...

    def veri_ekleme(eklenecek_veri, veri_seti):
        #Öncelikle eşleştirdiğimiz sütun isimleri atmak için boş bir değişken oluşturuyoruz ve eklenecek exceldeki sutun isimlerini daha rahat kullanabilmek için ayrı bir değişken yapıyoruz 
        new_columns = []
        columns = eklenecek_veri.columns
    
    
        # Columns döngüye sokup veri tabanındaki sütunlar ile  karşılaştırıp en yüksek oranı olanı elde ediyoruz ve onları new_columnslara ekliyoruz
        # Yanlış sütun isimleri vs girilmiş olabilir bu yüzden alakasız sütunlar eşleştirme yapmamak için 20 skoru yakalayamaz isek benzerlik bulanamadığını bildiriyoruz
        for i in columns:
            test = fuzz.WRatio(i, veri_seti.columns)
            if test <= 20:
                print("Yeterli benzerlik bulunmadı lütfen değişken isimlerini kontrol ediniz")
            else:
                new = max(veri_seti.columns, key = lambda name: fuzz.WRatio(i, name))
                new_columns.append(new)
    
        #Hatayı göz ardı edebilmek için try except içine alıyorum.
        try:
        
            #Elde ettiğimiz new columnsları eklenecek veri columnsları ile yer değiştiriyoruz ve artık veri setindeki sütun isimleri ile aynı stringleri elde etmiş oluyoruz.
            eklenecek_veri.columns = new_columns
    
            #Direk append metotu ile ekleyince indexlerde 0 1 0 1 gibi karmaşa oluyor onu çözmek için reindex değişkeni yaratıyoruz ve bu değişkende eklenecek veriyi veri setine ekliyoruz.
            #İndex sıralamasını düzeltmek için reset_index komutunu kullanıyoruz.
            #Reset index kullanılınca eski indexler index isimli yeni bir sütuna atanıyor ve bunu ortadan kaldırmak için o oluşan yeni sütunu siliyoruz.
            reindex = veri_seti.append(eklenecek_veri)
            reindex.reset_index(inplace=True)
            del reindex["İndex"]
    
            #En sonunda verileri yeni excel dosyasına kaydediyoruz ve eklenecek verileri doğru sütunlara ekleyiyoruz.
            reindex.to_excel("eklenmis_veri_tabani.xlsx")
    
            print("Veriler eklendi ve yeni excel dosyası oluşturuldu")
        except:
            pass 
    
    def bos_gozlem(self):
        a = self.veri.isnull().sum()  
        print(a)
        
    def bos_kaldir():
        pass
    
    def gozlum_doldur():
        pass
    

    

        
        
class gorsel():
    def __init__(self, x, y, veri):
        ...
        
    def kutu(x,y, veri):
        ax = sns.barplot(x, y, data= veri)
        
        return ax
    
    def hist():
        pass
    
    def sicaklık(): 
        pass
    
    def adet():
        pass
                       
```
